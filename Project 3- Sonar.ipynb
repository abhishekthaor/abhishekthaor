```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=36)
```

```python
import sklearn
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
model=[GaussianNB(), SVC(),DecisionTreeClassifier(),LogisticRegression()]
for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print("Accuracy score of ", m, "is:")
    print(accuracy_score(y_test,predm))
    print(confusion_matrix(y_test,predm))
    print(classification_report(y_test,predm))
    print('\n')
    
```

From the above results it is clear that with Accuracy score of 0.82 and f-1 score 0.85 SVC model is the best fit
